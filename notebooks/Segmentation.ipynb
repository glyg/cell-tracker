```python
default_path = '/media/data/xiaobo'
stack_io = get_stacks(default_path)
```

```python
magnification = 40
pixel_size = 6.45
stack_io.metadata['PhysicalSizeX'] = pixel_size / magnification
stack_io.metadata['PhysicalSizeY'] = pixel_size / magnification

stack_io.metadata['PhysicalSizeZ'] = 1.5
stack_io.metadata['TimeIncrement'] = 3.
```

```python
im0 = ct.inspect_stack(stack_io, stack_num=0, show_hist=True)
```

```python
def preprocess(stack):
    return stack / stack.max()
```

```python
stack_iterator = ct.build_iterator(stack_io, preprocess)
```

```python
parameters = {'segment_method':'naive', 'correction':0.1, 'min_z_size':1}
```

```python
cell_positions = nuclei_detector(stack_iterator(),
                                 metadata=stack_io.metadata,
                                 parameters=parameters,
                                 parallel=False,
                                 verbose=False,
                                 mapper=None,
                                 show_progress=False)
```

```python
fig, ax = plt.subplots()

ax.imshow(stack_io.get_tif().asarray().max(axis=0))
ax.plot(cell_positions.y, cell_positions.x, 'o')
```
